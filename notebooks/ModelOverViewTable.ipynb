```python

def plot_tlb():
    fig,ax = plt.subplots(figsize=(11.7, 8.3))
    rows = 9
    cols = 9
    ax.set_ylim(-1, rows + 3)
    ax.set_xlim(0, cols + .5)
    spacing = np.linspace(-1,rows+3,rows+1)
    
    for row in range(rows):
        d = table_info[row]
        mp = (spacing[row+1])- (spacing[row])
        y = spacing[row]+(mp/2)
        ax.text(x=.5, y=y, s=d['source_id'], va='center', ha='left')
        ax.text(x=1.5, y=y, s=d.get('Institution',''), va='center', ha='left')
        ax.text(x=4.0, y=y, s=d.get('Model resolution',''), va='center', ha='right')
        ax.text(x=5.3, y=y, s=d.get('Vertical levels',''), va='center', ha='right')
        ax.plot(
            [0, cols + 1.0],
            [y -.7, y - .7],
            ls=':',
            lw='.5',
            c='grey'
        )
    ax.text(.5,12.2,'Source id', weight='bold', ha='left')
    ax.text(1.5,12.2,'Institution', weight='bold', ha='left')
    ax.text(4.0,12.2,'Resolution \n (Lon \u00D7 Lat)', weight='bold', ha='right')
    ax.text(5.3,12.2,'Vertical\nlevels \n (model top)', weight='bold', ha='right')

#     for row in range(rows):
    ax.plot([0, cols + 1], [12, 12], lw='.5', c='black')
    ax.axis('off')

with mpl.rc_context(context_dict):
    plot_tlb()
```
